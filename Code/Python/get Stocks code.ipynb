## 종목코드, 종목명, 종류를 가져오는 크롤링 코드
- site : http://bigdata-trader.com/itemcodehelp.jsp
- BeautifulSoup 이용
- __맨 마지막 줄에 본인이 원하는 경로 지정 필수!__

```python
import urllib.request
from bs4 import BeautifulSoup
import urllib.request 
import pandas as pd
import numpy as np
```

```python
req = urllib.request.Request('http://bigdata-trader.com/itemcodehelp.jsp') 
data = urllib.request.urlopen(req).read() 
bs = BeautifulSoup(data, 'html.parser') ##받아온 url parsing 작업
td_Tag = bs.find_all('td') ## 그 중에 td태그만 가져옴
```

```python
len(td_Tag)/3
```

```python
allCode = []
for i in td_Tag:
    allCode.append(i.text)
```

```python
pd.DataFrame(np.array(allCode).reshape(2274,3))
```

```python
pd.DataFrame(np.array(allCode).reshape(2274,3)).to_csv('[본인경로]',index = False)
```
